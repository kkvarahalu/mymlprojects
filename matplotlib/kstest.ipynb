```python
import numpy as np  # Importing numpy for random number generation

di = [3.7, 7.9, 1.2, 4.5, 20.1]  # Given values

n_list = []  # To store normalized values
d_re = []    # To store cumulative sum values

si = sum(di)  # Step 1: Find total sum
# si = 3.7 + 7.9 + 1.2 + 4.5 + 20.1 = 37.4

# Step 2: Normalize each value
for i in di:
    n_di = i / si         # Example: 3.7 / 37.4
    n_list.append(n_di)   # Append to normalized list

# n_list = [0.0989, 0.2112, 0.0321, 0.1203, 0.5374]

# Step 3: Cumulative sum
for i in range(len(n_list)):
    if i == 0:
        d_re.append(n_list[i])         # First value is same
    else:
        d_re.append(d_re[i-1] + n_list[i])  # Add current with previous

# d_re = [0.0989, 0.3101, 0.3422, 0.4625, 1.0]

# Step 4: Generate a random number from uniform distribution
r = np.random.uniform(0, 1, 1)[0]  # Example: 0.6

# Step 5: Select value using proportional sampling
for i in range(len(d_re)):
    if r <= d_re[i]:
        print("Selected value:", di[i])
        break
```
