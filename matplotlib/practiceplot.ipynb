```python
x=[2,3,4,5,6]
from scipy.stats import anderson
a=anderson(x)
```

```python
a
```

```python
t=a.statistic
t
```

```python
a.critical_values[2]
```

```python
def ander(n):
    if n.statistic>n.critical_values[2]:
        print("non gaussian")
    else:
        print("gaussian")
```

```python
ander(a)
```
