```python
df=pd.read_csv("iris.csv")
```

```python
df
```

```python
sns.kdeplot(data=df,x=df['PetalLengthCm'],hue='Species')
```

```python
sample_mean=[]
sample_size=30
no_of_samples=1000
for i in range(no_of_samples):
    s=np.random.choice(df["PetalLengthCm"],size=sample_size,replace=True)
    sample_mean.append(np.mean(s))
```

```python
sample_mean
```

```python

sns.histplot(sample_mean,kde=True,color='red')
plt.title("CLT")
plt.xlabel("sampling distribution of samples mean")
plt.ylabel("count")
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
data=np.random.normal(loc=20,scale=5,size=10000)
stats.probplot(data,dist='norm',plot=plt)
plt.show()
```

```python
data=np.random.normal(loc=20,scale=5,size=50)
stats.probplot(data,dist='norm',plot=plt)
plt.show()
```

```python
data=np.random.uniform(low=-1,high=1,size=1000)
stats.probplot(data,dist='norm',plot=plt)
plt.show()
```

```python
import pylab as pylab
data=np.random.uniform(low=-1,high=1,size=1000)
stats.probplot(data,dist='norm',plot=pylab)
pylab.show()
```

```python
df=pd.read_csv("iris.csv")
```

```python
df
```

```python
data=df['PetalLengthCm']
stats.probplot(data,dist='norm',plot=pylab)
pylab.show()
```

```python
data=df['SepalLengthCm']
stats.probplot(data,dist='norm',plot=pylab)
pylab.show()
```

```python
data=df['SepalWidthCm']
stats.probplot(data,dist='norm',plot=pylab)
pylab.show()
```

```python
data=df['PetalWidthCm']
stats.probplot(data,dist='norm',plot=pylab)
pylab.show()
```

```python
import seaborn as sns
sns.kdeplot(data=df[df['Species']=='Iris-setosa'],x='SepalLengthCm',cumulative=True,label='Iris-setosa')
sns.kdeplot(data=df[df['Species']=='Iris-virginica'],x='SepalLengthCm',cumulative=True,label='Iris-virginica')
sns.kdeplot(data=df[df['Species']=='Iris-versicolor'],x='SepalLengthCm',cumulative=True,label='Iris-versicolor')
plt.title("CDF for SL")
plt.xlabel("sepal length")
plt.ylabel("probability P(X<=SL)")
plt.legend()
plt.show()
```
