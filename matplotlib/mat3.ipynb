```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("iris.csv")
```

```python
df
```

```python
df.columns
```

```python
df["Species"].value_counts()
```

```python
df.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
plt.grid()
```

```python
#sepal length is varying from 4 to 8
#sepal width is varying fron 1.8 to 4.5
```

```python
sns.FacetGrid(df,hue="Species").map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend();
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(df,hue="Species",height=4).map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend();
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(df,hue='Species',height=3,diag_kind='kde')
plt.show()
```

```python
sns.FacetGrid(df,hue='Species',height=3).map(sns.distplot,'SepalLengthCm')
plt.show()
```

```python
sns.FacetGrid(df,hue='Species',height=3).map(sns.kdeplot,'SepalLengthCm')
plt.show()
```

```python
sns.boxplot(x="Species",y="SepalLengthCm",data=df)
```
