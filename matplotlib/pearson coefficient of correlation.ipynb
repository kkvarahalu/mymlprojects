```python
import numpy as np
```

```python
h=np.array([140,160,155,170,163,145])
w=np.array([45,48,50,52,53,54])
cv=np.cov(h,w)
```

```python
cv
```

```python
cv[0,1]
```

```python
high_inch=h*0.4
weight_lbs=w*2.2
```

```python
cv2=np.cov(high_inch,weight_lbs)
```

```python
cv2
```

```python
cv2[0,1]
```

```python
cv3=np.corrcoef(high_inch,weight_lbs)
```

```python
cv3
```

```python
cv3[0,1]
```
