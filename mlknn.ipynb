```python
import numpy as np
import pandas as pd
```

```python
d=load_iris()
```

```python
df=pd.DataFrame(d.data,columns=d.feature_names)
```

```python
df
```

```python
df["target"]=d.target
```

```python
df
```

```python
df
```

```python
def species(x):
    if x==0:
        return 'setosa'
    elif x==1:
        return 'versicolor'
    else:
        return 'virginica'
df["variety"]=df["target"].apply(species)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
df
```

```python
X=df.drop(["target","variety"],axis='columns')
y=df["target"]
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
