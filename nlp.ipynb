```python
#basic unigram
from sklearn.feature_extraction.text import CountVectorizer
```

```python
d1="pasta is very tasty and affordable"
d2="pasta is not tasty and affordable"
bow=CountVectorizer(ngram_range=(1,1))
result=bow.fit_transform([d1,d2]).toarray()
print(result)
```

```python
def fun(d1,d2):
    bow=CountVectorizer(ngram_range=(3,3))
    result=bow.fit_transform([d1,d2]).toarray()
    print("Trigrams:",bow.get_feature_names_out())
    print("Result:\n",result)
    return result
fun(d1,d2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def fun(d1,d2):
    bow=TfidfVectorizer(ngram_range=(3,3))
    result=bow.fit_transform([d1,d2]).toarray()
    print("Trigrams:", bow.get_feature_names_out())
    print("Result:\n", result)
    return result
fun(d1,d2)
```
