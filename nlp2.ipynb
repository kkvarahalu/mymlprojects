```python
#above all transper to functions
#bag of words
from sklearn.feature_extraction.text import CountVectorizer
def fun(d1,d2):
    bow=CountVectorizer(ngram_range=(1,1))
    result=bow.fit_transform([d1,d2]).toarray()
    print("Vocabulary:",bow)
    print("Vocabulary:",bow.get_feature_names_out())#alfabetical sorted order d1,d2
    print("bow (unigram) Matrix:\n",result)
    return result
d1="pasta is very tasty and affordable"
d2="pasta is not tasty and affordable"
fun(d1,d2)
#Each row = sentence
#Each column = word count based on vocabulary
```

```python
#2.Bag of Words (Bigram → ngram_range=(2,2))
def fun(d1,d2):
    bow=CountVectorizer(ngram_range=(2,2))
    result=bow.fit_transform([d1,d2]).toarray()
    print("vocabulary(bigrams):",bow.get_feature_names_out())
    print("BoW (Bigram) Matrix:\n",result)                   
    return result
fun(d1,d2)
#Each bigram is a pair of 2 consecutive words.
```

```python
#3. Bag of Words (Trigram → ngram_range=(3,3))
def fun(d1,d2):
    bow=CountVectorizer(ngram_range=(3,3))
    result=bow.fit_transform([d1,d2]).toarray()
    print("vocabulary(trigrams):",bow.get_feature_names_out())#alfabetical sorted three words d1,d2
    print("BoW (Bigram) Matrix:\n",result)
    return result
fun(d1,d2)
#Each trigram is a group of 3 consecutive words.
```

```python
#4. TF-IDF (Unigram)
from sklearn.feature_extraction.text import TfidfVectorizer
def fun(d1,d2):
    bow=TfidfVectorizer()
    result=bow.fit_transform([d1,d2]).toarray()
    print("vocabulary(trigrams):",bow.get_feature_names_out())
    print("BoW (Bigram) Matrix:\n",result)
    return result
fun(d1,d2)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
#combination
def fun(d1,d2):
    bow=CountVectorizer()
    result=bow.fit_transform([d1,d2]).toarray()
    print("Bag of Words (BoW):")
    print(pd.DataFrame(result,columns=bow.get_feature_names_out()))
    tfidf=TfidfVectorizer()
    result1=tfidf.fit_transform([d1, d2]).toarray()
    print("tf-ide")
    print(pd.DataFrame(result1,columns=tfidf.get_feature_names_out()))
    return result
d1="pasta is very tasty and affordable"
d2="pasta is not tasty and affordable"
fun(d1,d2)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

def fun(d1, d2):
    # Bag of Words
    bow = CountVectorizer()
    result = bow.fit_transform([d1, d2]).toarray()
    print("Bag of Words (BoW):")
    print(pd.DataFrame(result, columns=bow.get_feature_names_out()))
    print()

    # TF-IDF
    tfidf = TfidfVectorizer()
    result1 = tfidf.fit_transform([d1, d2]).toarray()
    print(" TF-IDF:")
    print(pd.DataFrame(result1, columns=tfidf.get_feature_names_out()))
    return result

# Input texts
d1 = "pasta is very tasty and affordable"
d2 = "pasta is not tasty and affordable"

fun(d1, d2)
```
