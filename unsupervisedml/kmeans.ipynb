```python
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['target']=iris.target
```

```python
from sklearn.cluster import KMeans
```

```python
df=df.drop(['petal length (cm)','petal width (cm)'],axis=1)
```

```python
df
```

```python
df=df.drop('target',axis=1)
```

```python
df
```

```python
km=KMeans(n_clusters=3)
ypred=km.fit_predict(df)
```

```python
ypred
```

```python
df['clusters']=ypred
```

```python
df
```

```python
df1=df[df['clusters']==0]
df2=df[df['clusters']==1]
df3=df[df['clusters']==2]
```

```python
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],c='blue')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],c='red')
plt.scatter(df3['sepal length (cm)'],df3['sepal width (cm)'],c='black')
```

```python
cost_function_values=[]
k_values=range(1,15)
for i in k_values:
    km=KMeans(n_clusters=i)
    km.fit(df)
    cost_function_values.append(km.inertia_)
```

```python
cost_function_values
```

```python
plt.xlabel('k-values')
plt.ylabel('cost_function_values')
plt.title('Elbow_plot')
plt.plot(k_values,cost_function_values)
```
