```python
import numpy as np
def slr_closed(x, y):
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_xy = sum(x * y)
    sum_x2 = sum(x ** 2)

    b = ((n * sum_xy) - (sum_x * sum_y)) / ((n * sum_x2) - (sum_x ** 2))
    a = (sum_y - b * sum_x) / n

    return a, b

x = np.array([2, 4, 6, 8])
y = np.array([3, 7, 5, 10])
print(slr_closed(x, y))
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df
```

```python
df.isnull().sum()
```

```python
sns.boxplot(df['Height(Inches)'])
```

```python
sns.boxplot(df['Weight(Pounds)'])
```

```python
q1=df['Height(Inches)'].quantile(0.25)
q3=df['Height(Inches)'].quantile(0.75)
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
df=df[(df['Height(Inches)']>=lb)&(df['Height(Inches)']<=ub)]
print(df['Height(Inches)'])
```

```python
q1=df['Weight(Pounds)'].quantile(0.25)
q3=df['Weight(Pounds)'].quantile(0.75)
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
df=df[(df['Weight(Pounds)']>=lb)&(df['Weight(Pounds)']<=ub)]
print(df['Weight(Pounds)'])
```

```python
df
```

```python
sns.boxplot(df['Height(Inches)'])
```

```python
sns.boxplot(df['Weight(Pounds)'])
```

```python
df.columns
```

```python
df=df.drop("Index",axis=1)
```

```python
df
```

```python
x=df[['Weight(Pounds)']]
```

```python
x
```

```python
y=df[['Height(Inches)']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = lr.predict(x_test)
accuracy = accuracy_score(y_test.values.ravel(), y_pred.ravel())
print("Accuracy:", accuracy)
```

##    | Solver        | Use for            | Notes               |
      | ------------- | ------------------ | ------------------- |
      | `'lbfgs'`     | L2 only            | Fast, default       |
      | `'liblinear'` | L1 or L2           | Binary data ki best |
      | `'newton-cg'` | L2 only            | Large data          |
      | `'sag'`       | L2 only            | Large datasets      |
      | `'saga'`      | L1, L2, elasticnet | Works for all       |


```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,50,10,1,0.1]
solver=['lbfgs','liblinear','newton-cg','newton-cholesky','sag','saga']
```

```python
parameter=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import GridSearchCV
#you are importing GridSearchCV, a tool that automatically finds 
#  the best combination of hyperparameters for your model.
```

```python
grid=GridSearchCV(estimator=model,param_grid=parameter,scoring="accuracy",cv=5)
```

```python
grid(xtrain,ytrain)
```

```python
grid.best_param_
```

```python
grid.best_score_
```
